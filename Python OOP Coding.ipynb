```python
class GoldInvoice:
    def __init__(self,item_id,item_name,item_qty,item_rate,item_weight,item_pwc,item_pdis):
        self.item_id=item_id
        self.item_name=item_name
        self.item_qty=item_qty
        self.item_rate=item_rate
        self.item_weight=item_weight
        self.item_pwc=item_pwc
        self.item_pdis=item_pdis
def calc_item_price_exGST(goldInvoice_list,item_id_input):
    for i in goldInvoice_list:
        if(i.item_id == item_id_input):
            item_cost = (i.item_qty * i.item_weight * i.item_rate)
            pwc = item_cost * ( i.item_pwc / 100 )  #percentage of wastage
            pdis = item_cost * ( i.item_pdis / 100 ) #percentage of discharge
            total_cost = ( item_cost + pwc ) - pdis
        return total_cost
    return None    
def calc_item_price_inGST(goldInvoice_list,item_id_input):
    for i in goldInvoice_list:
        if(i.item_id == item_id_input):
            item_cost = (i.item_qty * i.item_weight * i.item_rate)
            pwc = item_cost * ( i.item_pwc / 100 )  #percentage of wastage
            pdis = item_cost * ( i.item_pdis / 100 ) #percentage of discharge
            gst = ( item_cost * (3 / 100 ) )
            total_cost =( ( item_cost + pwc ) - pdis ) + gst
        return total_cost  
    return None

if __name__ == '__main__':
    goldInvoice_list=[]
    goldInvoice_Count=int(input("goldInvoice_Count: "))
    for i in range(goldInvoice_Count):
        gold_item_id=int(input("gold_item_id: "))
        gold_item_name=input("gold_item_name: ")
        gold_item_qty=float(input("gold_item_qty: "))
        gold_item_rate=float(input("gold_item_rate: "))
        gold_item_weight=float(input("gold_item_weight: "))
        gold_item_pwc=float(input("gold_item_pwc: "))
        gold_item_pdis=float(input("gold_item_pdis: "))
        g=GoldInvoice(gold_item_id,gold_item_name,gold_item_qty,gold_item_rate,
                      gold_item_weight,gold_item_pwc,gold_item_pdis)
        goldInvoice_list.append(g)
    
    item_id_input=int(input("item_id_input: "))
    
    func1=calc_item_price_exGST(goldInvoice_list,item_id_input)
    if ( func1 == None ):
        print("No Item Found")
    else:
        print("{:.3f}".format(func1))
        
    func2=calc_item_price_inGST(goldInvoice_list,item_id_input)
    if ( func1 == None ):
        print("No Item Found")
    else:
        print("{:.2f}".format(func2))    
    
```

### CPA 25th FEB 2nd Slot (BLOOD)
#### (CPA) - CAMERA PROCTURED ASSESMENT

```python
class Blood:
    def __init__(self,bloodGroup,unitinHand):
        self.bloodGroup=bloodGroup
        self.unitinHand=unitinHand
        
class BloodBank():
    def __init__(self,blood_list):
        self.blood_list=blood_list
        
    def isBloodAvailable(self,blood_group_required,units_required):
        for i in self.blood_list:
            if (i.bloodGroup == blood_group_required and i.unitinHand>units_required):
                return True
        return None    
        
    def findMinBloodStock(self):
        units=[]
        for i in self.blood_list:
            units.append(i.unitinHand)
        min_unit=min(units)
        for i in self.blood_list:
            if i.unitinHand==min_unit:
                print(i.bloodGroup)
        

if __name__ == '__main__':
    blood_list=[]
    blood_count=int(input("Blood Count: "))
    for i in range(blood_count):
        blood_group=input("Blood Group: ")
        blood_group=blood_group.upper()
        unit_in_hand=int(input("unit in hand: "))
        b=Blood(blood_group,unit_in_hand)
        blood_list.append(b)
    blood_group_required=input("blood_group_required: ")
    blood_group_required=blood_group_required.upper()
    units_required=int(input("units_required: "))
    
    class1=Blood(blood_group,unit_in_hand)
    class2=BloodBank(blood_list)
    
    method1_class2=class2.isBloodAvailable(blood_group_required,units_required)
    if method1_class2 == None:
        print("Blood Not Available")
    elif method1_class2 == True:
        print("Blood Available")
    
    
    #method2_class2=class2.findMinBloodStock()
    class2.findMinBloodStock()    
```
